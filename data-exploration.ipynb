```python
import numpy as np
```

```python
books["missing_description"] = np.where(books["description"].isna(),1,0)
books["age_of_book"] = 2025 - books['published_year']
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman") #Non-continous values
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True,fmt=".2f", cmap="coolwarm",
                      cbar_kws ={"label": "Spearman Correlation"} )

heatmap.set_title("Correlation heatmap")

plt.show()
```

```python
books_missing =  books[~(books['description'].isna()) &
      ~(books['num_pages'].isna()) & 
      ~(books['average_rating'].isna())  &
      ~(books['published_year'].isna())]
```

```python
# Generate the value counts, reset index, and sort by 'count'
category_counts = books_missing['categories'].value_counts().reset_index().sort_values("count", ascending=False)

# Rename columns for better readability
category_counts.columns = ['Category', 'Count']

# Create the bar chart
plt.figure(figsize=(10,6))
plt.bar(category_counts['Category'], category_counts['Count'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Distribution of Missing Books')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"]>= 25]
```

```python
books_missing_25_words
```

New Feature creation :

 - We have title and subtitle books 
 - new field title and subtitle 

```python
books_missing_25_words['title_and_subtitle']= (
    np.where(books_missing_25_words['subtitle'].isna(),books_missing_25_words['title'],
             books_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
books_missing_25_words
```

New Description column :

- Combine description with isbn column

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join, axis=1)
books_missing_25_words.head()
```

Data is ready now drop the columns that we dont need

**Save the Cleaned Data in The File**

```python
(
    books_missing_25_words
    .drop(["subtitle","missing_description","words_in_description","age_of_book"], axis=1)
    .to_csv("books_cleaned.csv",index=False)
)
```
